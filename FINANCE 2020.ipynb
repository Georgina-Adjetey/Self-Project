```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
My_chat=pd.read_csv('Finance_Chat.txt', error_bad_lines=False, header=None, encoding='utf-8')
```

```python
My_chat
```

##### Categorizing Data

```python
My_chat.columns
```

```python
My_chat.columns=['Date','Chat','User','']
date=My_chat['Date']
date=date.str.split('[',n=1,expand=True)[1]
```

```python
date
```

##### Data Separation

```python
time_message=My_chat['Chat'].str.split(']',n=1,expand=True)
time=time_message[0]
time_message
```

```python
user_message=time_message[1].str.split(':',n=1,expand=True)
user_message
```

```python
time
```

```python
user=time_message[1].str.split
message=user_message[1]
user=user_message[0]
```

```python
message
```

```python
user
```

```python
MyChat=pd.DataFrame({'Date':date, 'Time':time, 'User':user, 'Message':message})
MyChat
```

```python
MyChat.dropna()
```

```python
MyChat.to_csv('Finance_Chat.csv',index=False)
```
